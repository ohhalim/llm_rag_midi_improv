```python
# 문제
# 두 정수 A와 B를 입력받은 다음, A+B를 출력하는 프로그램을 작성하시오.

# 입력
# 첫째 줄에 A와 B가 주어진다. (0 < A, B < 10)

# 출력
# 첫째 줄에 A+B를 출력한다.

# 예제 입력 1 
# 1 2
# 예제 출력 1 
# 3
```

```python
# map : 글자를 숫자로 바꿔주는거
# split :입력한 내용을 쪼개는거 ,공백을 기준으로
A, B = map(int, input().split())

print(A + B)
```

```python
# 문제
# 두 정수 A와 B를 입력받은 다음, A-B를 출력하는 프로그램을 작성하시오.

# 입력
# 첫째 줄에 A와 B가 주어진다. (0 < A, B < 10)

# 출력
# 첫째 줄에 A-B를 출력한다.

# 예제 입력 1 
# 3 2
# 예제 출력 1 
# 1
```

```python
# 문제
# 두 정수 A와 B를 입력받은 다음, A×B를 출력하는 프로그램을 작성하시오.

# 입력
# 첫째 줄에 A와 B가 주어진다. (0 < A, B < 10)

# 출력
# 첫째 줄에 A×B를 출력한다.

# 예제 입력 1 
# 1 2
# 예제 출력 1 
# 2
# 예제 입력 2 
# 3 4
# 예제 출력 2 
# 12
```

```python
A, B = map(int, input().split())

print(A / B) 
```

```python
# 문제
# 두 자연수 A와 B가 주어진다. 이때, A+B, A-B, A*B, A/B(몫), A%B(나머지)를 출력하는 프로그램을 작성하시오. 

# 입력
# 두 자연수 A와 B가 주어진다. (1 ≤ A, B ≤ 10,000)

# 출력
# 첫째 줄에 A+B, 둘째 줄에 A-B, 셋째 줄에 A*B, 넷째 줄에 A/B, 다섯째 줄에 A%B를 출력한다.

# 예제 입력 1 
# 7 3
# 예제 출력 1 
# 10
# 4
# 21
# 2
# 1
```

```python
A, B = map(int, input().split())
print(A + B) 
print(A - B) 
print(A * B) 
print(int(A / B))  
print(A % B) 

```
